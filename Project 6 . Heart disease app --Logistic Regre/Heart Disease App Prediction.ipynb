# Heart Disease App Prediction

```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv(r"C:\Users\ma516\OneDrive\Desktop\Machine Learning Projects\Heart disease app\heart_disease_data.csv")
data.tail(50)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['target'].unique()
```

```python
X = data.iloc[:,:-1]
Y = data["target"]
```

# Trained our Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_test , x_train , y_test ,y_train = train_test_split(X ,Y , test_size=0.2 , random_state= 3)
```

```python
print(X.shape , x_train.shape , y_train.shape)
```

# Using Logistic Regrssion Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter= 1000)
```

```python
model.fit(x_train , y_train )
```

```python
from sklearn.metrics import accuracy_score
```

```python
predict_data = model.predict(x_train)
accuracy_score_of_trained_data = accuracy_score(predict_data , y_train)
print("accuracy_score_of_trained_data" , accuracy_score_of_trained_data)
```

```python
predict_data = model.predict(x_test)
accuracy_score_of_trained_data = accuracy_score(predict_data , y_test)
print("accuracy_score_of_trained_data" , accuracy_score_of_trained_data)
```

```python
# input the variable
input_data = (63,1,3,145,233,1,0,150,0,2.3,0,0,1)
# Convert into array 
array_data = np.array(input_data)
# reshaped our data 
reshaped_data = array_data.reshape(1,-1)
#  predicted data
prediction = model.predict(reshaped_data)
print(prediction)
if(prediction == 0):
    print("Person Does not have a Heart Disease")
else:
    print("Person have Heart Disese")
```
