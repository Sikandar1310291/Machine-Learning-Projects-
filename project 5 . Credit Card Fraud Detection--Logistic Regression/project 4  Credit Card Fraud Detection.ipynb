```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\ma516\OneDrive\Desktop\Machine Learning Projects\Credit Card Fraud Detection\creditcard.csv")
data.tail()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
x  =  data.iloc[:,:-1]
y  =  data['Class'] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state=2 , test_size=0.2)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_train , y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_trained_data = model.predict(x_train)
accuracy_score_training_data = accuracy_score(x_trained_data,y_train)
print(accuracy_score_trained_data)
```

```python
x_testing_data = model.predict(x_test)
accuracy_score_testing_data = accuracy_score(x_testing_data , y_test)
print(accuracy_score_testing_data)
```
