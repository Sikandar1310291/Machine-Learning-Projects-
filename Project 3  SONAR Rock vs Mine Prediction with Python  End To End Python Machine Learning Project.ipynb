```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv(r"C:\Users\ma516\Downloads\Copy of sonar data.csv")
data.head()
```

```python
data["R"].unique()
```

```python
x = data.iloc[:,:-1]
y = data["R"]
```

```python
print(x)
print(y)
```

# Training Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state= 2 , test_size=0.2)
```

# Using  Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

# Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
Y_predict_data = model.predict(x_train)
Accuracy_Score_of_predict_data = accuracy_score(y_train , Y_predict_data)
print(Accuracy_Score_of_predict_data)
```
