# Import Libraires

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Load The Dataset
data = pd.read_csv(r"C:\Users\ma516\OneDrive\Desktop\Machine Learning Projects\Air Quality Prediction App\AirQualityData.csv")
data.head()
```

```python
data.shape
```

```python
#  Make Column for x and y
x_data = data.drop(columns=["AirQualityIndex" , 'Date' , "Time" , "DayOfWeek" , "Hour"] , axis=1)
y_data = data[['AirQualityIndex']]
```

```python
x_data.head()
```

```python
y_data.head()
```

```python
#  Standardize Our Data
ss = StandardScaler()
```

```python
x = ss.fit_transform(x_data)
print(x[:5])
```

```python
# Trained our model
x_train , x_test , y_train ,y_test = train_test_split(x,y_data,random_state=2,test_size=0.2)
```

```python
print(x.shape , x_train.shape , x_test.shape , y_train.shape , y_test.shape)
```

```python
model = KNeighborsRegressor()
```

```python
#  fit our model
model.fit(x_train , y_train)
```

```python
# check the best fit n_neighbors four data
for i in range(1 , 30 ,  ):
    model = KNeighborsRegressor( n_neighbors= i)
    model.fit(x_train , y_train)
    print(i , model.score(x_train , y_train) ,model.score(x_test , y_test) )
```
