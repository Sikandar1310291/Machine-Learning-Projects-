#  Titanic Survival app  Prediction

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\ma516\OneDrive\Desktop\Machine Learning Projects\Project 15 Titanic Survival Prediction\train.csv")
data.head()
```

```python
data.isnull().sum()
```

# Fill The Missing Value

```python
mean_age  = data["Age"].mean()
print(mean_age)
data['Age'].fillna(mean_age , inplace= True)
```

```python

data["Cabin"].fillna(method="bfill" , inplace=True)
data["Embarked"].fillna(method="bfill" , inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

# Encode our Values for Training 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['Name'] =le.fit_transform( data['Name'])
data['Sex'] =le.fit_transform( data['Sex'])
data['Cabin'] =le.fit_transform( data['Cabin'])
data['Embarked'] =le.fit_transform( data['Embarked'])
data['Ticket'] =le.fit_transform( data['Ticket'])
```

```python
data.head()
```

```python
x = data.drop("Survived" , axis= 1)
y = data["Survived"]
```

```python
print(x.shape)
print(y.shape)
```

```python
print(x,y)
```

```python
print(x.ndim , y.ndim)
```

# Trained Our Model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test  = train_test_split(x, y , random_state=42 ,test_size=0.2)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# Using LogisticRegression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

# Find The Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
Trained_data = model.predict(x_train)
accuracy_score_of_trained_data = accuracy_score(Trained_data , y_train)
print(accuracy_score_of_trained_data)
```

```python
Testng_data = model.predict(x_test)
accuracy_score_of_tested_data = accuracy_score(Testng_data , y_test)
print(accuracy_score_of_tested_data)
```

# Check the Prediction of our Model

```python
input_data = ( 1    ,   3  , 108  ,  1 , 22.000000  ,    1 ,     0  ,   523  , 7.2500 ,81     ,    2)
arrayed_data = np.array(input_data)
reshaped_data = arrayed_data.reshape(1,-1)
prediction = model.predict(reshaped_data)
print(prediction)
if(prediction == 0 ):
    print("Person is Died")
else:
    Print("Person is Survived")
```
