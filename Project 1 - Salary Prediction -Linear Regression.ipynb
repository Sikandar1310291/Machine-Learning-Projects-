```python
------------------------------------------------------------------Linear Regression--------------------------------------------------------------------

Y = wX + b

Y --> Dependent Variable

X --> Independent Variable

w --> weight

b --> bias

Gradient Descent:

Gradient Descent is an optimization algorithm used for minimizing the loss function in various machine learning algorithms. It is used for updating the parameters of the learning model.

w = w - α*dw

b = b - α*db

Learning Rate:

Learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.
```

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\ma516\Downloads\salary_data.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x = "YearsExperience" , y = "Salary" , data = data)
plt.show()
```

```python
x = data[['YearsExperience']]
y = data[['Salary']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train , y_train )
```

```python
input_data = input("Write Your Experience in Year :")

input_list = [float(x) for x in input_data.split(",")]

array_data = np.asarray(input_list)

reshaped_data = array_data.reshape(1,-1)

predicted_data = model.predict(reshaped_data)
print(predicted_data)
```
