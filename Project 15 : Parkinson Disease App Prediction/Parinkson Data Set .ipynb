```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"C:\Users\ma516\OneDrive\Desktop\Machine Learning Projects\Parinkson Prediction App\parkinsons.data")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
x  = data.drop(columns=["status" , "name"])
y = data["status"]
```

```python
pd.set_option('display.max_columns', None)
x.tail()
```

```python
y.tail()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model = SVC(kernel="linear")
```

```python
model.fit(x_train , y_train )
```

```python
trained_data =  model.predict(x_train)
accuracy_score_of_trained_data = accuracy_score(trained_data , y_train)
print("accuracy_score_of_trained_data :" , accuracy_score_of_trained_data)
```

```python
tested_data =  model.predict(x_test)
accuracy_score_of_tested_data = accuracy_score(tested_data , y_test)
print("accuracy_score_of_tested_data :" , accuracy_score_of_tested_data)
```

```python
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```
